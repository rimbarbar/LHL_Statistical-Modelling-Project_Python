```python
# Query Foursquare API
def query_foursquare(lat, lon, radius=1000, categories="13065,13003", retries=5, backoff_factor=0.5):
    params = {
        "ll": f"{lat},{lon}",
        "radius": radius,
        "categories": categories,
        "limit": 50
    }
    retries = int(retries)
    for attempt in range(retries):
        try:
            response = requests.get(FOURSQUARE_URL, headers=headers, params=params, timeout=10)
            response.raise_for_status()  # Raise an exception for HTTP errors
            return response.json()  # Return JSON response on success
        except ChunkedEncodingError as e:
            print(f"ChunkedEncodingError encountered. Retrying... ({attempt + 1}/{retries})")
            time.sleep(backoff_factor * (2 ** attempt))  # Exponential backoff
        except RequestException as e:
            print(f"Request failed for {lat}, {lon}: {e}")
            break  # Exit on non-recoverable errors

    return None  # Return None if all retries fail
```

```python
def parse_foursquare(data):
    results = []
    if data and "results" in data:
        for poi in data["results"]:
            results.append({
                "name": poi.get("name"),
                "category": poi["categories"][0]["name"] if poi.get("categories") else None,
                "address": poi["location"]["formatted_address"] if "location" in poi else None,
                "lat": poi["geocodes"]["main"]["latitude"] if "geocodes" in poi else None,
                "lon": poi["geocodes"]["main"]["longitude"] if "geocodes" in poi else None
            })
    return results
```

```python
def query_yelp(lat, lon, radius=1000, categories="restaurants,bars"):
    params = {
        "latitude": lat,
        "longitude": lon,
        "radius": radius,
        "categories": categories,
        "limit": 50
    }
    response = requests.get(YELP_URL, headers=YELP_HEADERS, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Yelp API Error: {response.status_code}")
        return None
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
def parse_yelp(data):
    results = []
    if data and "businesses" in data:
        for business in data["businesses"]:
            results.append({
                "name": business.get("name"),
                "category": business["categories"][0]["title"] if business.get("categories") else None,
                "address": ", ".join(business["location"]["display_address"]),
                "rating": business.get("rating"),
                "lat": business["coordinates"]["latitude"] if "coordinates" in business else None,
                "lon": business["coordinates"]["longitude"] if "coordinates" in business else None
            })
    return results
```
